```python
(
    gapminder
    .query('year == 2007')
    .groupby ('continent')
    .agg({'lifeExp': 'mean'})
    .sort_values('lifeExp', ascending=False)
)
```

Changing shape of data

```python
#transferred from tidy to wide format
(
    gapminder
    .filter(['continent', 'country', 'lifeExp', 'year'])
    .pivot(columns='year',
          values='lifeExp',
          index=['continent', 'country'])
)
```

```python
#lets make wide format tidy again
(
    gapminder
    .filter(['continent', 'country', 'lifeExp', 'year'])
    .pivot(columns='year',
          values='lifeExp',
          index=['continent', 'country'])
    .reset_index()
  # do not touch continent and country columns and remind that everyth else is life expectancy 
    .melt(id_vars=['continent', 'country'],
         value_name='lifeExp')
)
```
